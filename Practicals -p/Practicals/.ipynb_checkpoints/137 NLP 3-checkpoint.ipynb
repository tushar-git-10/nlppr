```python
def viterbi_algorithm(obs,states,pi,trans,emit):
    v = [{}]
    path = {}

    #Initialization
    for state in states:
        v[0][state] = pi[state] * emit[state][obs[0]]      ## obs[0] means 1 row
        path[state] = [state]

    best_state = max(v[0],key=v[0].get)
    print("After 1 word (",obs[0],") -> ",best_state)

    for t in range(1,len(obs)):
        v.append({})
        new_path = {}

        for curr_state in states:
            max_prob = 0
            best_prev = None

            for prev_state in states:
                prob = (v[t-1][prev_state] * trans[prev_state][curr_state] * emit[curr_state][obs[t]])

                if prob > max_prob:
                    max_prob = prob
                    best_prev = prev_state

            v[t][curr_state] = max_prob
            new_path[curr_state] = path[best_prev] + [curr_state]

        path = new_path

        best_state = max(v[t],key=v[t].get)
        print("After word ",t+1,"(",obs[t],") -> ",best_state)
        print(v)

    final_state = max(v[-1],key=v[-1].get)
    final_seq = path[final_state]

    print("\n Final Best Sequence:",final_seq)
```

```python
states = ['N', 'V']
obs = ['time', 'flies']

pi = {'N': 0.6, 'V': 0.4}

trans = {
    'N': {'N': 0.3, 'V': 0.7},
    'V': {'N': 0.8, 'V': 0.2}
}

emit = {
    'N': {'time': 0.5, 'flies': 0.2},
    'V': {'time': 0.4, 'flies': 0.6}
}

viterbi_algorithm(obs, states, pi, trans, emit)
```
