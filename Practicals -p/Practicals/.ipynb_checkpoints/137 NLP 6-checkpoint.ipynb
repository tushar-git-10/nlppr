```python
import nltk
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet as wn

sentence1 = "I went to the bank to deposit money in my account"
tokens1 = word_tokenize(sentence1)

sense1 = lesk(tokens1, "bank", pos=wn.NOUN)

print("Sentence 1:", sentence1)
print("Predicted Sense:", sense1)
print("Definition:", sense1.definition())


sentence2 = "The river water overflowed the bank after heavy rain"
tokens2 = word_tokenize(sentence2)

sense2 = lesk(tokens2, "bank", pos=wn.NOUN)

print("Sentence 2:", sentence2)
print("Predicted Sense:", sense2)
print("Definition:", sense2.definition())
```

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def improved_lesk(sentence, word):
    
    context = set(word_tokenize(sentence.lower()))
    context = context - stop_words
    
    best_sense = None
    max_overlap = 0
    
    # Consider only noun meanings
    for sense in wn.synsets(word, pos=wn.NOUN):
        
        # Gloss = definition + examples 
        signature = word_tokenize(sense.definition().lower())
        
        for example in sense.examples():
            signature += word_tokenize(example.lower())
        
        signature = set(signature) - stop_words
        
        # Count overlap
        overlap = len(context.intersection(signature))
        
        if overlap > max_overlap:
            max_overlap = overlap
            best_sense = sense
    
    return best_sense
```

```python
s1 = "I went to the bank to deposit money in my account"
s2 = "The river overflowed the bank after heavy rain"

sense1 = improved_lesk(s1, "bank")
sense2 = improved_lesk(s2, "bank")

print("Sentence 1:", s1)
print("Predicted Sense:", sense1)
print("Definition:", sense1.definition())

print("Sentence 2:", s2)
print("Predicted Sense:", sense2)
print("Definition:", sense2.definition())
```
