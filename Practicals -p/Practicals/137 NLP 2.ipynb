## With Backtracking

```python
def min_edit(str1,str2):
    m = len(str1)
    n = len(str2)

    T = []
    for i in range(m+1):
        row = [] 
        for j in range(n+1):
            row.append(0)
        T.append(row)

    for i in range(m+1):
        for j in range(n+1):

            # For null
            if i==0 and j==0:
                T[i][j] = 0

            # for first row
            elif i == 0:
                T[i][j] = T[i][j-1] + 1

            # for first Col
            elif j == 0:
                T[i][j] = T[i-1][j] + 1

            #if row == col 
            elif str1[i-1] == str2[j-1]:
                T[i][j] = T[i-1][j-1]

            # if Row != Col
            else:
                T[i][j] = min(T[i][j-1], T[i-1][j], T[i-1][j-1]) + 1

    i, j = m, n
    operations = []

    while i > 0 or j > 0:

        # Match
        if i > 0 and j > 0 and str1[i-1] == str2[j-1]:
            operations.append("Match")
            i -= 1
            j -= 1

        # Replace
        elif i > 0 and j > 0 and T[i][j] == T[i-1][j-1] + 1:
            operations.append("Replace")
            i -= 1
            j -= 1

        # Delete
        elif i > 0 and T[i][j] == T[i-1][j] + 1:
            operations.append("Delete")
            i -= 1

        # Insert
        else:
            operations.append("Insert")
            j -= 1

    operations.reverse()

    return T[m][n], operations, T
```

```python
dist, ops, matrix = min_edit("intention", "execution")

for row in matrix:
    print(row)
print("Edit Distance:", dist)
print("Edit Operations Only:")
print("Replace:", ops.count("Replace"))
print("Insert :", ops.count("Insert"))
print("Delete :", ops.count("Delete"))
```
