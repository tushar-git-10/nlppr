```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
x = df.drop("quality",axis=1)
y = df['quality']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x = sc.fit_transform(x)
```

```python
from sklearn.cluster import AgglomerativeClustering
ag = AgglomerativeClustering(n_clusters=2)
```

```python
cluster = ag.fit_predict(x)
```

```python
df['cluster'] = cluster
```

```python
df['cluster'].value_counts()
```

```python
df.groupby('cluster')['quality'].mean()      ## High quality is Red wine(max mean) and low quality is White wine(min mean)
```

### DENDOGRAM

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
link = linkage(x,method='ward')    ## method for the finding the distance between two clusters
```

```python
plt.figure(figsize=(16, 6))
dendrogram(link)
plt.title("Dendrogram")
plt.xlabel("Cluster Size")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=df['cluster'], cmap='plasma')
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Clusters")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(x, df['cluster'])
print("Silhouette Score:", score)
```
