```python
# Process 1: Input Text
para = "Artificial Intelligence is the simulation of human intelligence by computer systems, enabling machines to learn, reason, solve problems, and perceive their environment. It utilizes algorithms and data to perform tasks like language translation, image recognition, and decision-making."
print(para)
```

```python
# Process 2: Word Tokenization
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize
tokens = word_tokenize(para)
print(tokens)
```

```python
# Process 3: Sentence Tokenization
from nltk.tokenize import sent_tokenize
sent = sent_tokenize(para)
print(sent)
```

```python
# Process 4: Lowercase Conversion
lower = []
for word in tokens:
    lower.append(word.lower())
print(lower)
```

```python
# Process 5: Remove Punctuation
import string
no_punct = []
for word in lower:
    if word not in string.punctuation:
        no_punct.append(word)
print(no_punct)
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
filtered = []

for word in no_punct:
    if word not in stop_words:
        filtered.append(word)

print(filtered)
```

```python
# Process 7: Stemming
from nltk.stem import PorterStemmer
ps = PorterStemmer()

for word in filtered:
    print(word, "->", ps.stem(word))
```

```python
# Process 8: Lemmatization
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()

for word in filtered:
    print(word, "->", lemma.lemmatize(word))
```

```python
# Process 9: Final Output
final = ""
for word in filtered:
    final = final + " " + word

print(final)
```
