```python
# Process 1: Input Words
s1 = "kitten"
s2 = "sitting"
```

```python
# Process 2: MED + Operations (Simple Version)

def med(str1, str2):

    m = len(str1)
    n = len(str2)

    T = []
    for i in range(m+1):
        row = []
        for j in range(n+1):
            row.append(0)
        T.append(row)

    for i in range(m+1):
        for j in range(n+1):

            if i == 0 and j == 0:
                T[i][j] = 0

            elif i == 0:
                T[i][j] = T[i][j-1] + 1

            elif j == 0:
                T[i][j] = T[i-1][j] + 1

            elif str1[i-1] == str2[j-1]:
                T[i][j] = T[i-1][j-1]

            else:
                T[i][j] = min(T[i][j-1], T[i-1][j], T[i-1][j-1]) + 1

    i, j = m, n
    ops = []

    while i > 0 or j > 0:

        if i > 0 and j > 0 and str1[i-1] == str2[j-1]:
            ops.append("Match")
            i -= 1
            j -= 1

        elif i > 0 and j > 0 and T[i][j] == T[i-1][j-1] + 1:
            ops.append("Replace")
            i -= 1
            j -= 1

        elif i > 0 and T[i][j] == T[i-1][j] + 1:
            ops.append("Delete")
            i -= 1

        else:
            ops.append("Insert")
            j -= 1

    ops.reverse()
    return T[m][n], ops, T
```

```python
# Process 3: Show Matrix

dist, ops, matrix = med(s1, s2)

for r in matrix:
    print(r)
```

```python
# Process 4: Result

print("MED =", dist)
print("Replace:", ops.count("Replace"))
print("Insert :", ops.count("Insert"))
print("Delete :", ops.count("Delete"))
```
