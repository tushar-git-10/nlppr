```python
sentence = ["I", "love", "coding"]
tags = ["PRON", "VERB", "NOUN"]
transition_prob = {
    "START": {"PRON": 0.6, "VERB": 0.2, "NOUN": 0.2},
    "PRON": {"VERB": 0.7, "NOUN": 0.2, "PRON": 0.1},
    "VERB": {"NOUN": 0.6, "VERB": 0.3, "PRON": 0.1},
    "NOUN": {"VERB": 0.4, "NOUN": 0.4, "PRON": 0.2}
}
emission_prob = {
    "PRON": {"I": 0.9, "love": 0.1, "coding": 0.0},
    "VERB": {"I": 0.0, "love": 0.8, "coding": 0.2},
    "NOUN": {"I": 0.0, "love": 0.1, "coding": 0.9}
}
```

```python
V = [{}]
path = {}

# Initialization
for tag in tags:
    V[0][tag] = transition_prob["START"][tag] * emission_prob[tag][sentence[0]]
    path[tag] = [tag]

# Recursion
for t in range(1, len(sentence)):
    V.append({})
    new_path = {}

    for curr in tags:
        max_prob = -1
        best_prev = None

        for prev in tags:
            prob = V[t-1][prev] * transition_prob[prev][curr] * emission_prob[curr][sentence[t]]
            if prob > max_prob:
                max_prob = prob
                best_prev = prev

        V[t][curr] = max_prob
        new_path[curr] = path[best_prev] + [curr]

    path = new_path

# Find best final tag
best_tag = max(V[-1], key=V[-1].get)
best_sequence = path[best_tag]
```

```python
print("Sentence:", sentence)
print("POS Tags:", best_sequence)
```
