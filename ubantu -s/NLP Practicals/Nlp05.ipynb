```python
# Import Natural Language Toolkit
import nltk

# Tokenization + POS tagging tools
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag

# Download required datasets
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
# Sample text for co-reference resolution
text = """
John went to the market. He bought apples.
Mary saw John there. She greeted him.
The dog was barking loudly. It was hungry.
"""
```

```python
# Split text into sentences
sentences = sent_tokenize(text)
sentences
```

```python
# Gender-based name lists (simple rule-based system)

male_names = ['john', 'ram', 'rohit', 'rahul']
female_names = ['mary', 'sita', 'priya', 'anita']

# Store detected entities
persons, objects = [], []
```

```python
print("\nPOS TAGGING:\n")

for s in sentences:
    tagged = pos_tag(word_tokenize(s))
    print(tagged)

    for w, t in tagged:
        w = w.lower()

        if t == "NNP":
            persons.append(w)
        elif t == "NN":
            objects.append(w)

print("\nCO-REFERENCE LINKS:\n")
```

```python
for s in sentences:
    tagged = pos_tag(word_tokenize(s))

    for w, t in tagged:
        w = w.lower()

        if t == "PRP":
            if w in ["he", "him", "his"]:
                for x in reversed(persons):
                    if x in male_names:
                        print(w, "→", x)
                        break

            elif w in ["she", "her"]:
                for x in reversed(persons):
                    if x in female_names:
                        print(w, "→", x)
                        break

            elif w == "it" and objects:
                print(w, "→", objects[-1])
```
