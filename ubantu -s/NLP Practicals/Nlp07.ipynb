```python
import nltk
from nltk.corpus import brown
from gensim.models import Word2Vec, FastText

nltk.download('brown')
```

```python
sentences = brown.sents()

print("Total sentences:", len(sentences))
print("Sample:", sentences[0])
```

```python
sentences = [[word.lower() for word in sent] for sent in sentences]
```

```python
w2v_model = Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=5,
    workers=1
)
```

```python
print("Similar to 'good':")
print(w2v_model.wv.most_similar("good"))

print("\nSimilarity (man vs woman):")
print(w2v_model.wv.similarity("man", "woman"))
```

```python
ft_model = FastText(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=5,
    workers=1
)
```

```python
print("\nFastText Similar to 'good':")
print(ft_model.wv.most_similar("good"))

print("\nFastText similarity (run vs running):")
print(ft_model.wv.similarity("run", "running"))
```
