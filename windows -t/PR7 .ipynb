```python
# Word Embedding using Word2Vec (Simple)

from gensim.models import Word2Vec

# Sample sentences (training data)
sentences = [
    ["i", "love", "machine", "learning"],
    ["machine", "learning", "is", "fun"],
    ["i", "love", "coding"],
    ["coding", "is", "fun"]
]

# Train Word2Vec model
model = Word2Vec(sentences, vector_size=50, window=2, min_count=1, workers=1)

# Get word vector
print("Vector for 'machine':\n", model.wv["machine"])

# Similar words
print("\nWords similar to 'learning':")
print(model.wv.most_similar("learning"))

# Similarity score
print("\nSimilarity between 'coding' and 'learning':")
print(model.wv.similarity("coding", "learning"))
```
